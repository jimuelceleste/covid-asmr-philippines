## Exploratory Data Analysis

What provinces from COVID-19 dataset are not in the current population data?

```python
# 2020 Philippine Population Data from PSA
df = pd.read_excel("data/Philippine Population 2020 by Province and Age Group.xlsx", index_col=0)
df = df.T
df.shape
```

```python
# Columns: Provinces
df.columns
```

```python
# Provinces from the COVID-19 Data
mortality = pd.ExcelFile("output/mortality.xlsx")
provinces = mortality.sheet_names
provinces
```

```python
# Provices in COVID-19 data that are not in population data
for p in provinces:
    if p not in df.columns:
        print(p)
```

```python
# Provinces in population data that are not in COVID-19 data
for p in df.columns:
    if p not in provinces:
        print(p)
```

All provinces from COVID-19 data are in the population data. The next step is to calculate the mortality rates.
