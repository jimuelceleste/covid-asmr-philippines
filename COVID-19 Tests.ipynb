## COVID-19 Tests Performed per Facility

````python
tests
````

````python
tests[tests['facility_name']=='A Star Laboratories']
````

Filtering out records with report_date > 2023-05-09. The COVID-19 mortality data for this study were collected only until this date. 

````python
tests['report_date'] = pd.to_datetime(tests['report_date'])
tests
````

````python
tests = tests[tests['report_date'] <= pd.to_datetime(' 2023-05-09')]
tests
````

Taking the max "cumulative_samples_tested" per facility (group by facility_name).

https://scales.arabpsychology.com/stats/how-to-find-the-max-value-by-group-in-pandas/

````python
tests_df = tests.groupby('facility_name')['cumulative_samples_tested'].max().reset_index()
tests_df
````

````python
tests_df = tests_df.rename(columns={
    'facility_name': 'Facility', 
    'cumulative_samples_tested': 'Cumulative Samples Tested'
})
tests_df
````

````python
tests_df['Facility'] = tests_df['Facility'].apply(lambda x: x.lower())
tests_df
````

````python
output_file = "/Users/jimuelcelestejr/Documents/codebook/covid-asmr-philippines/data/COVID-19 Tests/covid19_tests_performed.csv"
tests_df.to_csv(output_file, index=False)
````
